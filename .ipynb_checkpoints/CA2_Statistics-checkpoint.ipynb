## T-test Two Populations

If we want to see the difference in hours worked between male and female clerical workers.

```python
#Grouping data according to the categories

male = df1.query('Type_of_Worker == "Clerical employees - male"')['value']
female = df1.query('Type_of_Worker == "Clerical employees - female"')['value']
```

```python
#Assuming H0 that mu1 = mu2 (Clerical employees - male = Clerical employees - female)
test = stats.ttest_ind(male, female, 
                      equal_var=True)

display(test)
```

```python
#Assuming H0 that mu1 = mu2 (Clerical employees - male = Clerical employees - female)
test = stats.ttest_ind(male, female, 
                      equal_var=False)

display(test)
```

Due to the extremely small pvalue we can reject H0 and confirm average hours worked by the male employes is not equal to the average hours worked by female employees.

### DF10 Number of People Employed in Construction by Country

#### Two sample T-test

```python
from scipy.stats import ttest_ind

groups = df10.groupby('LOCATION')

# Extract data for each group
data1 = groups.get_group('IRL')['Value']
data2 = groups.get_group('NZL')['Value']

ttest_ind(data1, data2, equal_var=True)
```

By setting equal_var = True, a standard two sample t-test is conducted. This considers the population variance to be equal.

#### Welches T-test

```python
ttest_ind(data1, data2, equal_var=False)
```

By setting equal_var = False, this changes the test to 'Welches T-test' which does not consider the populations to have equal variance. Despite having different equal_var specified the results remained similar. Both t-tests reject the null hypothesis.

#### Paired T-Test

```python
groups = df10.groupby('LOCATION')

# Extract data for each group
data1 = groups.get_group('IRL')['Value']
data2 = groups.get_group('NZL')['Value']
  
# Performing the paired sample t-test
stats.ttest_rel(data1, data2)
```

The paired samples t-test follows the null and alternative hypotheses:

H0: It signifies that the mean number of construction workers ireland and New Zealand are equal.
HA: It signifies that the mean number of construction workers ireland and New Zealand are not equal.
As the p-value comes out to be equal to 0.000138 which is less than 0.05, we reject the null hypothesis. So, we have enough proof to claim that the true mean number of construction workers is different for Ireland and New Zealand.

#### Kruskal-Wallis H-test

```python
from numpy.random import seed
from numpy.random import randn
from scipy.stats import kruskal

groups = df10.groupby('LOCATION')

# Extract data for each group
data1 = groups.get_group('IRL')['Value']
data2 = groups.get_group('AUS')['Value']
data3 = groups.get_group('NZL')['Value']

# Compute the Kruskal-Wallis H-test
stat, p = st.kruskal(data1, data2, data3)

# Print the test statistics and p-value
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p > alpha:
 print('Same distributions (fail to reject H0)')
else:
 print('Different distributions (reject H0)')
```

Ref - https://machinelearningmastery.com/nonparametric-statistical-significance-tests-in-python/

#### Mann Whitney U Test

Non parametric significance test used to determine whether two independant variables were drawn from a population within the same distribution. Typically works best with a sample size greater than 20 observations.

```python
from scipy.stats import mannwhitneyu

groups = df10.groupby('LOCATION')

# Extract data for each group
data1 = groups.get_group('IRL')['Value']
data2 = groups.get_group('NZL')['Value']

# compare samples
stat, p = mannwhitneyu(data1, data2)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
```

## Binomial 

Assuming the probability of a construction worker working more than the mean average hours worked in a given week, take p=0.5. Checking the hours of 50 workers (sample size = n = 50). What is the probability 20 of them will be working more than average? 

```python
from scipy.stats import binom
# setting the values
# of n and p
n = 50
p = 0.5
# defining the list of r values
r_values = list(range(n + 1))
# obtaining the mean and variance 
mean, var = binom.stats(n, p)
# list of pmf values
dist = [binom.pmf(r, n, p) for r in r_values ]
# printing the table
print("r\tp(r)")
for i in range(n + 1):
    print(str(r_values[i]) + "\t" + str(dist[i]))
# printing mean and variance
print("mean = "+str(mean))
print("variance = "+str(var))
#Ref - https://www.geeksforgeeks.org/python-binomial-distribution/
```
